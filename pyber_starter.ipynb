```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
citydata_df = pd.read_csv(city_data_to_load)
ridedata_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(citydata_df,ridedata_df,how='outer',on='city')

# Display the data table for preview
combined_df.head(3)
```

## Bubble Plot of Ride Sharing Data

```python
cities = combined_df.groupby("city")
avgfare = cities.mean()["fare"]
ridecount = cities["ride_id"].count()
drivercount = cities.max()["driver_count"]
citytype = citydata_df.set_index("city")["type"]
city_table_df = pd.DataFrame({
    "Number of Rides": ridecount,
    "Average Fare": avgfare,
    "Driver Count": drivercount,
    "City Type": citytype
})

urban = city_table_df[city_table_df['City Type']=="Urban"]
suburban = city_table_df[city_table_df['City Type']=="Suburban"]
rural = city_table_df[city_table_df['City Type']=="Rural"]
```

```python
urbanridecount = urban["Number of Rides"]
urbanfares = urban["Average Fare"]
urbandrivercount = urban["Driver Count"]
suburbanridecount = suburban["Number of Rides"]
suburbanfares = suburban["Average Fare"]
suburbandrivercount = suburban["Driver Count"]
ruralridecount = rural["Number of Rides"]
ruralfares = rural["Average Fare"]
ruraldrivercount = rural["Driver Count"]

plt.scatter(urbanridecount, urbanfares, s=urbandrivercount*10, color="red",
            edgecolor="black", label="Urban",alpha=.8, linewidths=.25)
plt.scatter(suburbanridecount, suburbanfares, s=suburbandrivercount*10, color="blue",
            edgecolor="black", label="Suburban",alpha=.8,linewidths=.25)
plt.scatter(ruralridecount, ruralfares, s=ruraldrivercount*10, color="yellow",
            edgecolor="black", label="Rural",alpha=.8,linewidths=.25)

plt.suptitle("Pyber Ride Sharing Data (2016)", fontsize=14)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

lgnd = plt.legend(title='City Types')

plt.text(1, 0.5, "Note: Circle size correlates with driver count per city.", fontsize=10, transform=plt.gcf().transFigure)
plt.savefig("Pyber_PlotterBubble.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares = combined_df.sum()['fare']
urban_data = combined_df[combined_df['type']=="Urban"]
suburban_data = combined_df[combined_df['type']=="Suburban"]
rural_data = combined_df[combined_df['type']=="Rural"]

urbantotalfare = urban_data['fare'].sum()
suburbantotalfare = suburban_data['fare'].sum()
ruraltotalfare = rural_data['fare'].sum()
farevalues = [urbantotalfare,ruraltotalfare,suburbantotalfare]

# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
colors = ["green", "orange", "lightblue"]
plt.pie(farevalues,explode=(0.1,0,0),colors=colors,startangle=285,labels=labels,autopct="%1.1f%%",shadow=True)
plt.suptitle("% of Total Fares by City Type", fontsize=14, fontweight='bold')
plt.axis("equal")

# Save Figure
plt.savefig("Pyber_FarePie.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides = combined_df.count()['ride_id']
urban_ride_count = urban_data['ride_id'].count()
suburban_ride_count = suburban_data['ride_id'].count()
rural_ride_count = rural_data['ride_id'].count()

# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
colors = ["green", "orange", "lightblue"]
ridevalues = [urban_ride_count,rural_ride_count,suburban_ride_count]
plt.pie(ridevalues,explode=(0.1,0,0),colors=colors,startangle=265,labels=labels,autopct="%1.1f%%",shadow=True)
plt.suptitle("% of Total Rides by City Type", fontsize=14, fontweight='bold')

# Save Figure
plt.savefig("Pyber_RidesPie.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
totaldrivers = citydata_df.groupby('type').sum()['driver_count']
labels = ["Rural","Suburban","Urban"]
colors = ["green", "orange", "lightblue"]

# Build Pie Charts
plt.pie(totaldrivers,explode=(0,0,0.1),startangle=155,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
plt.suptitle("% of Total Drivers by City Type", fontsize=14, fontweight='bold')

# Save Figure
plt.savefig("Pyber_DriversPie.png")
```
